```python
#adapted decons_tg code 
def load_tg_tables(pattern):
    return (
        pd.read_csv(pattern, sep="\t")
            .drop(["forward_from_message_id", "via_bot_id", "reply_to", "link_url", "grouped_message_id", "is_action_type", "message_object_record"], axis=1)
            .assign(
                record_id=lambda df: "msg_" + df["peer_id"].astype(str) + "_" + df["message_id"].astype(str)
            )
    )
    
def merge_tg_tsv_tables(pattern):
    mg = load_tg_tables(pattern)
    mg = mg.copy()[mg.message_text.notna()]
    # because of the variable dtypes in forward_from_channel, we have to force it to string. there should be no need for the user to work with this column for NLP anyway.
    
    mg["forward_from_channel"] = mg["forward_from_channel"].fillna("None").astype(str)
    output_path = os.path.join(pq_path, f"{channel}.pqt")
    mg.to_parquet(output_path, index=False)
    print(f"Database {channel} saved to {output_path}.")

    
    return mg

```

```python
#run through before set channel_names.
pq_path = 'S:/ERP Raw Data/pien/THESIS/parquets'
for channel in channel_names:
    current_directory = 'S:/ERP Raw Data/pien'
    os.chdir(current_directory)
    
    pattern = f"{channel}_table.tsv"
    print("Loading Telegram TSV tables...")
    mg = merge_tg_tsv_tables(pattern)
    print(mg.head())
```
